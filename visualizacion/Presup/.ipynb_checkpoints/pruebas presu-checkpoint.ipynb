```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.float_format', '{:.2f}'.format)

def secciones (n):
    if n < 10:
        x = '0'+str(n)
    else:
        x =str(n)
    return 'S'+x

def apli_presup(apli):
    aplicacion1 =apli.split('-')
    descripcion = aplicacion1[1]
    aplicacion2 =aplicacion1[0]
    aplicacion = aplicacion2.split()
    programa =aplicacion[1]
    concepto =aplicacion[2]
    
    return programa, concepto, descripcion
    
```

```python
# letcura del fichero excel mensual
meses =['ENERO','FEBRERO','MARZO','ABRIL','MAYO','JUNIO','JULIO','AGOSTO','SEPTIEMBRE','OCTUBRE']

df_final = pd.DataFrame()
for m in meses:
    xls = pd.ExcelFile(rf'C:\Users\ragod\OneDrive\Escritorio\Proyectos_TB\Proyectos_TB\visualizacion\Presup\datos\MENSUAL {m} 2021 ANEXO I.XLSX')
    hojas = xls.sheet_names
    for s in hojas:
        if s != 'S5  21':
            df1 = pd.read_excel(xls, s)
            df1['Mes']= m
            df1['Seccion']= s
            df_final=pd.concat([df_final,df1],axis = 0)
            
            
df_final = df_final.reset_index(drop=True)
df_final.columns=['Organismo','Aplicacion','Credito_Inicial','Credito_Final','Obligaciones','Mes','Seccion']
```

```python
df_final
```

```python
#limpieza
borrar = df_final[df_final['Aplicacion']=='TOTAL SERVICIO'].index

# Delete these row indexes from dataFrame
df_final.drop(borrar , inplace=True)

borrar = df_final[df_final['Aplicacion']=='APLICACIÓN PGE'].index
df_final.drop(borrar , inplace=True)

borrar = df_final[df_final['Aplicacion'].isnull()].index
df_final.drop(borrar , inplace=True)

borrar = df_final[df_final['Aplicacion'].str.contains(' -')== False].index
df_final.drop(borrar , inplace=True)
```

```python
df_final = df_final.reset_index(drop=True)
```

```python
df_final["Credito_Inicial"].fillna(0, inplace = True) 
df_final["Credito_Final"].fillna(0, inplace = True) 
df_final["Obligaciones"].fillna(0, inplace = True)
df_final["Credito_Inicial"].replace('-',0, inplace = True) 
df_final["Credito_Final"].replace('-',0, inplace = True) 
df_final["Obligaciones"].replace('-',0, inplace = True)  
```

```python
df_final[['Credito_Inicial','Credito_Final','Obligaciones']] = df_final[["Credito_Inicial",'Credito_Final','Obligaciones']].astype(float)
df_final
```

```python
df_final.to_csv('df_final.csv', index = False)
```
