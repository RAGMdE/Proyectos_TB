```python
import pandas as pd
import numpy as np
import fitz
```

```python
doc = fitz.open(r'C:/Users/ragod/OneDrive/Escritorio/2021_sin_selccionar.pdf') 
#lectura d ela pagina de clasificaci
lista =[]
for p in range(0,273):
    pagina = doc[p]
    paras = pagina.get_text('blocks')
    # la informaci√≥n empieza del bloque 13  al 33 apartir del elemento 5
    for i in range(14, len(paras)-1):
        lista.append(paras[i][4])
```

```python
lista[7558][18:21]
```

```python
lista2=[]
for n in range(0, 7600):
    lista2.append(lista[n])
```

```python

for n in range(len(lista2)):
    xx = lista2[n][18:21]
    if xx == '200':
        lista2[n]=lista2[n][:18]+'*\n*\n'+lista2[n][18:]
```

```python
for n in range(len(lista2)):
    xx = lista2[n][18:21]
    if xx == '000':
        lista2[n]=lista2[n][:18]+'*\n'+lista2[n][18:]
```

```python

for n in range(len(lista2)):
    xx = lista2[n][36:39]
    if xx == '200':
        lista2[n]=lista2[n][:36]+'*\n'+lista2[n][36:]

```
