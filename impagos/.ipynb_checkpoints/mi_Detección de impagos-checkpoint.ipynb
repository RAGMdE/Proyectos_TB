## 2. Extracción de variables

De las tablas **loan**, **account** y **district** extrae las siguientes variables en una única tabla
- identificador de la cuenta
- fecha del préstamo
- cantidad del préstamo
- duración del préstamo
- pagos mensuales
- estado del préstamo
- frecuencia de los extractos bancarios
- fecha de creación de la cuenta
- nº de habitantes del distrito
- variables del distrito (A4, A11, A12, A13, A14, A15, A16)

```python
query = '''
SELECT loan.account_id,
       loan.date AS date_loan,
       amount,
       duration,
       payments,
       status,
       frequency,
       account.date AS date_acc,
       A4, A11, A12, A13, A14, A15, A16

FROM loan
JOIN account
ON loan.account_id = account.account_id
JOIN district
ON account.district_id = district.district_id
'''
df = pd.read_sql(query,db)
df
```

Transforma las fechas con `pandas.to_datetime`

```python
df['date_loan'] = pd.to_datetime(df.date_loan,format='%Y-%m-%d')
df['date_acc'] = pd.to_datetime(df.date_acc,format='%Y-%m-%d')
```

Crea la variable `days_between` como la diferencia de días entre la fecha del préstamo y la fecha creación de la cuenta

```python
df['days_between'] = (df.date_loan-df.date_acc).dt.days
```

Renombra las variables relacionadas con el distrito y crea la variable del ratio de crímenes por habitante

```python
df['n_inhabitants'] = df.A4
df['average_salary'] = df.A11
df['average_unemployment_rate'] = df[['A12', 'A13']].mean(axis=1)
df['entrepreneur_rate'] = df['A14']
df['average_crime_rate'] = df[['A15','A16']].mean(axis=1) / df['n_inhabitants']
```

Crea el target binario

```python
df['Target'] = (df['status']=='B') | (df['status']=='D')
```

```python
df
```

De la tabla **trans** obtén, para cada cuenta, la cantidad transferida y el balance

```python
query = '''
SELECT trans.account_id, trans.amount as trans_amount, balance as trans_balance 
FROM trans
JOIN loan
ON trans.account_id = loan.account_id
AND trans.date < loan.date
'''
df_trans = pd.read_sql(query,db)
df_trans
```

Cuenta el número de transacciones de cada cuenta

```python
# account_id | n_trans
#    2          54
```

```python
# account_id | n_trans
#    2          54
n_trans = df_trans[['account_id','trans_amount']].groupby('account_id',
                                                         as_index=False).count()
n_trans.columns=['account_id','n_trans']
n_trans
```

Calcula la media de cantidad y balance de las transacciones para cada cuenta

```python
df_trans2 = df_trans.groupby('account_id',as_index=False).mean()
df_trans2
```

Une todos estos datos con el dataframe general

```python
df = df.merge(df_trans2,how='left',
              left_on='account_id',right_on='account_id').merge(n_trans,how='left',
                                                               left_on='account_id',
                                                               right_on='account_id')
df
```

De la tabla **card**, añade el tipo de tarjeta para cada cuenta (solo titulares)

```python
query = '''
SELECT disp.account_id, card.type as card_type
FROM card
JOIN disp
ON card.disp_id = disp.disp_id
JOIN loan
ON disp.account_id = loan.account_id
AND card.issued < loan.date
WHERE disp.type= 'OWNER'
'''
df_card = pd.read_sql(query,db)
df_card
```

Une la tabla anterior a la general. Para aquellas cuentas sin tarjeta, ponle el valor "No"

```python
df = df.merge(df_card, how='left', left_on='account_id',right_on='account_id')
df['card_type'].fillna('No',inplace=True)
df
```

De la tabla **client**, obtén la edad de los clientes (en el momento del préstamo), su sexo y añade una variable binaria que indique si el distrito del cliente coincide con el distrito de la cuenta

```python
query = '''
SELECT account_id,
       loan.date as loan_date,
       account.district_id as acc_dist_id,
       client.district_id as cli_dist_id,
       gender as owner_gender,
       birth_date

FROM loan
JOIN account using(account_id)
JOIN disp using(account_id)
JOIN client using(client_id)

WHERE disp.type = 'OWNER'
'''
df_cli = pd.read_sql(query,db)
df_cli
```

```python
df_cli['same_district'] = df_cli['acc_dist_id'] == df_cli['cli_dist_id']
df_cli['loan_date'] = pd.to_datetime(df_cli['loan_date'],format='%Y-%m-%d')
df_cli['birth_date'] = pd.to_datetime(df_cli['birth_date'],format='%Y-%m-%d')
df_cli['owner_age'] = (df_cli.loan_date-df_cli.birth_date).dt.days//365
df_cli
```

Une la tabla anterior con la general

```python
df = df.merge(df_cli,left_on='account_id',right_on='account_id')
df
```

## 6. Modelado

Entrena uno o varios modelos, dividiendo el conjunto en train-test 

Intentamos desarrollar un modelo de clasificacion. Se realziaran varias tecnicas. Buscamos ante todo precision: RF, SVM y boosting

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.compose import make_column_selector

# Se estandarizan las columnas numéricas y se hace one-hot-encoding de las 
# columnas cualitativas. Para mantener las columnas a las que no se les aplica 
# ninguna transformación se tiene que indicar remainder='passthrough'.
numeric_cols = X_sm.select_dtypes(include=['float64', 'int64','int32']).columns.to_list()
cat_cols = X_sm.select_dtypes(include=['object', 'category']).columns.to_list()

preprocessor = ColumnTransformer(
                  [('scale', StandardScaler(), numeric_cols),
                   ('onehot', OneHotEncoder(handle_unknown='ignore'), cat_cols)],
                remainder='passthrough')
X_train_prep = preprocessor.fit_transform(X_sm)
X_test_prep  = preprocessor.transform(X_test)
lista_seleccion=[]
```

```python
#DTC
from sklearn.tree import DecisionTreeClassifier as DTC
dtc=DTC(criterion='gini')

dtc.fit(X_train_prep, y_sm)

y_pred=dtc.predict(X_test_prep)

from sklearn.metrics import accuracy_score as f1s
train_score=dtc.score(X_train_prep, y_sm)
test_score=dtc.score(X_test_prep, y_test)
#acc(y_pred, y_test)
#lista_seleccion.append(['DTC',[train_score,test_score]])
'train Acc: ', train_score, '---  test Acc: ', test_score
```

```python
#SVM
from sklearn.svm import SVC

svc=SVC()
svc.fit(X_train_prep, y_sm)

y_pred=svc.predict(X_test_prep)

from sklearn.metrics import accuracy_score as acc

# acc(y_pred, y_test)
train_score=svc.score(X_train_prep, y_sm)
test_score=svc.score(X_test_prep, y_test)
lista_seleccion.append(['SVC',[train_score,test_score]])
'train Acc: ', train_score, '---  test Acc: ', test_score

```

```python
# Boosting
from xgboost import XGBClassifier as XGBC

xgbc=XGBC(use_label_encoder=False, n_estimators=10)

xgbc.fit(X_train_prep, y_sm)

train_score=xgbc.score(X_train_prep, y_sm)
test_score=xgbc.score(X_test_prep, y_test)
lista_seleccion.append(['XGBC',[train_score,test_score]])
'train Acc: ', train_score, '---  test Acc: ', test_score
```

```python
# KNNC
from sklearn.neighbors import KNeighborsClassifier as KNNC
knnc=KNNC()
knnc.fit(X_train_prep, y_sm)

train_score=knnc.score(X_train_prep, y_sm)
test_score=knnc.score(X_test_prep, y_test)
y_pred=knnc.predict(X_test_prep)
acc(y_pred, y_test)
lista_seleccion.append(['KNNC',[train_score,test_score]])


'train Acc: ', train_score, '---  test Acc: ', test_score
```

```python
# Random Forest
from sklearn.ensemble import RandomForestClassifier as RFC
rand_forest=RFC()
rand_forest.fit(X_train_prep, y_sm)
train_score=rand_forest.score(X_train_prep, y_sm)
test_score=rand_forest.score(X_test_prep, y_test)
lista_seleccion.append(['RFC',[train_score,test_score]])
'train Acc: ', train_score, '---  test Acc: ', test_score
```

```python
# Regresion logistica
from sklearn.linear_model import LogisticRegression as LOGREG

logreg=LOGREG()

logreg.fit(X_train_prep, y_sm)
train_score=logreg.score(X_train_prep, y_sm)
test_score=logreg.score(X_test_prep, y_test)
lista_seleccion.append(['log_reg',[train_score,test_score]])
'train Acc: ', train_score, '---  test Acc: ', test_score

```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.feature_selection import SelectKBest
from sklearn.model_selection import GridSearchCV
```

```python
grid_random_forest = {"n_estimators": [120],
                     "max_depth": [3,4,5,6,10,15,17],
                     "max_features": ["sqrt", 3, 4]                          
                     }


svc_param = {
            'C': [0.01, 0.1, 0.3, 0.5, 0.9], 
            'kernel': ['linear',"poly","rbf"],
            'coef0': [-10.,-1., 0., 0.1, 0.5, 1, 10, 100],
            'gamma': ('scale', 'auto')
            }


logreg_param = {    
                 "penalty": ['l1','l2'], 
                 "C": np.logspace(-3, 3, 7),
                 'solver'  : ['liblinear']
                }
xgbc_param ={'colsample_bytree': [0.5],
             'gamma': [0.2],
             'max_depth': [3],
             'min_child_weight': [3],
             'reg_alpha': [0.001],
             'subsample': [0.9]
            }
             
```

```python
# Almaceno en una lista de tuplas los modelos (nombre que le pongo, el modelo, hiperparametros)
models = [('rand_forest', rand_forest, grid_random_forest),
          ('xgbc', xgbc, xgbc_param),
          ('reg_log', logreg, logreg_param),
          ('svc', svc, svc_param),
         ]

# Declaro en un diccionario los pipelines e hiperparametros
models_gridsearch = {}

for i in models:
    models_gridsearch[i[0]] = GridSearchCV(i[1],
                                          i[2],
                                          cv = 10,
                                          verbose=1,
                                          n_jobs=-1,
                                          scoring='f1')
    
    models_gridsearch[i[0]].fit(X_train_prep, y_sm)
```

```python
best_grids = [(i, j.best_score_) for i, j in models_gridsearch.items()]

best_grids = pd.DataFrame(best_grids, columns=["Grid", "Best score"]).sort_values(by="Best score", ascending=False)
best_grids
```

```python
# El mejor ha sido rand_forest. Ya esta entrenada con todo train
models_gridsearch['rand_forest'].best_estimator_
```

```python
# La probamos en test
models_gridsearch['rand_forest'].best_estimator_.score(X_test, y_test)
```
