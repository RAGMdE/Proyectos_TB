```python
import pandas as pd
import xml.etree.ElementTree as ET
import ftplib
```

```python
ftp =ftplib.FTP('f31-preview.runhosting.com','4009006_DATOS','Rafa9999')
```

```python
filename ='OptaF24.xml'

ftp.retrbinary("RETR " + filename, open(filename, 'wb').write)
  
ftp.quit()
```

```python
tree = ET.parse('OptaF24.xml')
root = tree.getroot()
```

```python
for child in root:
    print(child)
```

```python
for element in child:
    print(element)
```

```python
#Craecion de una lista por cada columna
team=[]
half = []
minute =[]
second=[]
x_origin =[]
y_origin =[]
x_destination = []
y_destination =[]
outcome =[]
```

```python
for subelement in child:
    if subelement.attrib.get('type_id') =="1":
        if subelement.attrib.get('team_id')=='43':
            team.append('Manchester City')
        else:
            team.append('Bolton Wanderers')
        half.append(subelement.attrib.get('period_id'))
        minute.append(subelement.attrib.get('min'))
        second.append(subelement.attrib.get('sec'))
        outcome.append(subelement.attrib.get('outcome'))
        x_origin.append(subelement.attrib.get('x'))
        y_origin.append(subelement.attrib.get('y'))
        for q in subelement:
            if q.attrib.get('qualifier_id')=='140':
                x_destination.append(q.attrib.get('value'))
            if q.attrib.get('qualifier_id')=='141':
                y_destination.append(q.attrib.get('value'))       
        
```

```python
df = pd.DataFrame(list(zip(team,
                           half, 
                            minute, 
                            second,
                            x_origin, 
                            y_origin, 
                            x_destination,
                            y_destination, 
                            outcome)),
                       columns=['team','half','minute','second','x_origin','y_origin',
                                  'x_destination', 'y_destination','outcome'])
```

```python
df[['x_origin', 'y_origin','x_destination','y_destination']] = df[['x_origin', 'y_origin','x_destination','y_destination']].astype(float)
```

```python
df['x_distance']=df['x_destination']-df['x_origin']
```
